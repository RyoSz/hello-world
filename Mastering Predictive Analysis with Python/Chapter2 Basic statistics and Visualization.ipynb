<p>genreという新しい列を作りたい。</p>
<p>現状、当てはまるジャンルごとに、その列に1を入れてあり、その他は0をいれる複数行というような作り。</p>
<p>idxmaxを使い、「指定された列たちの中で、最初に最大値が来た列のIndexを返す」。()内のArg指定は、0=行で、1=列で返す。</p>

```python
imdb_rating['genre'] = imdb_rating[['Action','Animation','Comedy','Drama','Documentary','Romance']].idxmax(1)
```

```python
imdb_rating['genre'].head()
```

下のやり方は、辞書型といい、その作り方は下のように記述して宣言する。

```python
genres_map = {"Action":'red', "Animation":'blue', "Comedy":'yellow', "Drama":'green', "Documentary":'orange', "Romance":'purple'}
```

```python
print(genres_map)
```

```python
imdb_rating['genre_color'] = imdb_rating['genre'].apply(lambda x: genres_map[x])
```

```python
imdb_rating['genre_color'].head()
```

Pivotもできる。下の例は、「imdb_rating」というデータベースの中で、「rating」を、「year」でまとめあげて、「genre」別に配置したもの。まとめ方としてはmeanを使っている。

```python
pd.pivot_table(imdb_rating, values='rating', index='year', columns='genre', aggfunc=np.mean)
```

RuntimeをGenre毎に比較したい。その場合、同じRuntimeのデータポイントをまとめて、ヒストグラムをGenre毎に複数作るという手もあるが、少し見にくいであろう。<br>
なので、ここではヒストグラムを確率密度に置き換える。その手法としてKernel Density Estimation (KDE)を使う。
以下、1行目で「Genreでグルーピングして」「Runtimeデータを」「KDEで確率密度に置き換えて」「X軸の長さを0~120にして」「Genreの凡例をつけて」描画している。

```python
plot1 = imdb_rating.groupby('genre')['length'].plot(kind='kde', xlim=(0,120), legend='genre')
plot1[0].set_xlabel('Number of minutes')
plot1[0].set_title('Distribution of films by runtime minutes')
```

Ratingに置き換えたばあい。

```python
plot2 = imdb_rating.groupby('genre')['rating'].plot(kind='kde', xlim=(0,10), legend='genre')
```

以下Box plotでやる場合。Box plotでやる場合は、どうやらまずPivotでまとめるみたい。<br>

```python
pd.pivot_table(imdb_rating, values='rating', index='title', columns='genre').plot(kind='box', by='genre').set_ylabel('Rating')
```

alphaの意味が分からない。

```python
from pandas.tools.plotting import scatter_matrix
scatter_matrix(imdb_rating[['year', 'budget', 'length', 'rating', 'votes']], alpha=0.2, figsize=(10,10), diagonal='kde')
```

下もちょっとよくわからないところがあるけど、小難しくなっているのは、いっぺんに6つのScatterを表示させようとしているから。<br>
例えば、一番最後の「row+=1」がないと、6つの箱はできるけど、Scatterは一番左上の箱に全部まとまって表示されて、その他5つの箱はからっぽ。

```python
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(15,15))
row=0
col=0
for index, genre in imdb_rating.groupby('genre'):
    if row>2:
        row=0
        col+=1
    genre.groupby('genre').plot(ax=axes[row,col], kind='scatter', x='length', y='rating', s=np.sqrt(genre['votes']), c=genre['genre_color'], xlim=(0,120), ylim=(0,10), alpha=0.5, label=index)
    row+=1
```
